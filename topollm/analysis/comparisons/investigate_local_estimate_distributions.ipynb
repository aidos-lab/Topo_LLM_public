```python
def pick_range_of_local_estimate_values(df: pd.DataFrame, a: float, b: float) -> pd.DataFrame:
    df_copy = df.copy()
    df_copy = df_copy[(df_copy["local_estimate"] > a) & (df_copy["local_estimate"] < b)]
    return df_copy


df_base_model_lower_values = pick_range_of_local_estimate_values(df_first_model, a=10.4, b=10.6)
df_base_model_higher_values = pick_range_of_local_estimate_values(df_first_model, a=12.0, b=12.2)

df_finetuned_model_lower_values = pick_range_of_local_estimate_values(df_second_model, a=9.0, b=9.6)
df_finetuned_model_higher_values = pick_range_of_local_estimate_values(df_second_model, a=10.8, b=11.0)
```
